```python
# Creating a dictionary to impute Country-specific columns
country, abbr, tee, pop, unemp, urb_pop, lat, long = {}, {}, {}, {}, {}, {}, {}, {} 
for i in df['Country'].dropna().unique():
    df_sub = df[df['Country'] == i].iloc[0]
    
    country[i] = df_sub['Country']
    abbr[i] = df_sub['Abbreviation']
    tee[i] = df_sub['Gross tertiary education enrollment (%)']
    pop[i] = df_sub['Population']
    unemp[i] = df_sub['Unemployment rate']
    urb_pop[i] = df_sub['Urban_population']
    lat[i] = df_sub['Latitude']
    long[i] = df_sub['Longitude']
```

```python
# Defining a Function to easily impute missing values in Country-specific features
def Imputer(n, Country):
    df['Country'].loc[n] = country[Country]
    df['Abbreviation'].loc[n] = abbr[Country]
    df['Gross tertiary education enrollment (%)'].loc[n] = tee[Country]
    df['Population'].loc[n] = pop[Country]
    df['Unemployment rate'].loc[n] = unemp[Country]
    df['Urban_population'].loc[n] = urb_pop[Country]
    df['Latitude'].loc[n] = lat[Country]
    df['Longitude'].loc[n] = long[Country]
```

```python
# 'category' and 'channel_type' have a relationship. Making use of it, creating a dictionary to map
channel_category_dict = {'Music': 'Music', 'Games': 'Gaming', 
                         'Entertainment': 'Entertainment', 
                         'Education': 'Education', 
                         'People': 'People & Blogs', 
                         'Sports': 'Sports', 
                         'Animals': 'Pets & Animals', 
                         'Tech': 'Science & Technology', 
                         'News & Politics': 'News', 
                         'Film': 'Film & Animation', 
                         'Howto': 'Howto & Style', 
                         'Nonprofit': 'Nonprofits & Activism', 
                         'Comedy': 'Comedy'}

category_channel_dict = {j: i for i, j in channel_category_dict.items()}
category_channel_dict['Shows'] = 'Entertainment'
```

```python
for i in range(len(df)):

    if (pd.isna(df['category'].loc[i])) and (not pd.isna(df['channel_type'].loc[i])):
        df['category'].loc[i] = channel_category_dict[df['channel_type'].loc[i]]

    elif (pd.isna(df['channel_type'].loc[i])) and (not pd.isna(df['category'].loc[i])):
        df['channel_type'].loc[i] = category_channel_dict[df['category'].loc[i]]
```
